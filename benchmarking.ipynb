```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as pp

%matplotlib widget
```

# Scaling of combingation array with number of basis AOs

```python
def words_of_mem(N, unit='b'):
    if unit == 'b':
        scale = 1
    else: 
        scale = 8
    
    raw = 3 * N ** 4

    return raw / scale


    
```

```python
fig, ax = pp.subplots()

x = np.arange(0, 100)
y = words_of_mem(x, unit='B')

ax.semilogy(x, y)
```
